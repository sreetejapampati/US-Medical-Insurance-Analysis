```python
import csv
```

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []

def load_data(lst, column_name):
    with open('insurance.csv', 'r') as insurance_file:
        reader = csv.DictReader(insurance_file)
        for row in reader:
            lst.append(row[column_name])
        return lst

load_data(age, 'age')
load_data(sex, 'sex')
load_data(bmi, 'bmi')
load_data(children, 'children')
load_data(smoker, 'smoker')
load_data(region, 'region')
load_data(charges, 'charges')

```

```python
def insurance_dictionary(age,sex,bmi,children,smoker,region,charges):
    insurance_dictionary = dict()
    for i in range(len(age)):
        insurance_dictionary[i] = {"Age":age[i], "Sex":sex[i], "BMI":bmi[i], "Children":children[i], "Smoker":smoker[i], "Region":region[i], "Charges":charges[i]}
    return insurance_dictionary
insurance_dictionary = insurance_dictionary(age, sex, bmi, children, smoker, region, charges)
print(insurance_dictionary)

```

```python
class BasicInsuranceInfo:
    def __init__(self, age, sex, bmi, children, smoker, region, charges):
        self.age = age
        self.sex = sex
        self.bmi = bmi
        self.children = children
        self.smoker = smoker
        self.region = region
        self.charges = charges
        
    def average_age(self):
        total_age = 0
        for age in self.age:
            total_age = total_age + int(age)
        average_age = total_age/len(self.age)
        return "The average age of all the patients is {}.".format(average_age)
    

    def population_by_location(self):
        unique_locations = []
        for location in self.region:
            if location not in unique_locations:
                unique_locations.append(location) 
        num_southwest = 0
        num_southeast = 0
        num_northeast = 0
        num_northwest = 0
        for region in self.region:
            if region == "southwest":
                num_southwest +=1
            elif region == "southeast":
                num_southeast +=1
            elif region == "northeast":
                num_northeast +=1
            elif region == "northwest":
                num_northwest +=1   
        region_by_population = {"southwest":num_southwest, "southeast":num_southeast, "northeast":num_northeast, "northwest":num_northwest}
        return region_by_population

    def smoker_count_analysis(self):
        total_smokers = 0
        total_nonsmokers = 0
        for smoker_status in self.smoker:
            if smoker_status == "yes":
                total_smokers = total_smokers + 1
            elif smoker_status == "no":
                total_nonsmokers = total_nonsmokers + 1
        percent_smokers = 100*total_smokers/(total_smokers+total_nonsmokers)  
        percent_nonsmokers = 100*total_nonsmokers/(total_smokers+total_nonsmokers)
        return "Smokers consist of {percent_smokers}% of all covered by this insurance, whereas, non-smokers consist of {percent_nonsmokers}.".format(percent_smokers=percent_smokers, percent_nonsmokers = percent_nonsmokers)
     
    
                                        
#     def insurance_cost_if_smoker(self):
#         ages_for_smoker = []   
#         for smoker in self.smoker:
#             for age in self.age:
#                 if smoker=="yes":
#                     ages_for_smoker.append(age)
#         smoker_total_age = 0            
#         for age in ages_for_smoker:
#             smoker_total_age = smoker_total_age + int(age)
#         average_age_smoker = smoker_total_age/len(ages_for_smoker)   
#         return "The average age of all smokers in our insurance database is {}.".format(average_age_smoker)
# #create dictionaries - look at hurricane analysis
    

        
```

```python
smoker_charges_values = []
for charge in df.loc[df["smoker"]=="yes"].charges:
    smoker_charges_values.append(charge)
print(smoker_charges_values)    


    
    
```

```python
smoker_bmi_values = []
for value in df.loc[df["smoker"]=="yes"].bmi:
    smoker_bmi_values.append(value)
print(smoker_bmi_values)    


```

```python
smoker_region_values = []
for value in df.loc[df["smoker"]=="yes"].region:
    smoker_region_values.append(value)
print(smoker_region_values)    
```

```python
smoker_children_values = []
for value in df.loc[df["smoker"]=="yes"].children:
    smoker_children_values.append(value)
print(smoker_children_values)
```

```python
smoker_age_values = []
for value in df.loc[df["smoker"]=="yes"].age:
    smoker_age_values.append(value)
print(smoker_age_values)
```

```python
smoker_sex_values = []
for value in df.loc[df["smoker"]=="yes"].sex:
    smoker_sex_values.append(value)
print(smoker_sex_values)
```

```python
class SmokerAnalysis():
    def __init__(self, smoker_age_values, smoker_sex_values, smoker_bmi_values, smoker_children_values, smoker_region_values, smoker_charges_values):
        self.smoker_age_values = smoker_age_values
        self.smoker_sex_values = smoker_sex_values
        self.smoker_bmi_values = smoker_bmi_values
        self.smoker_children_values = smoker_children_values
        self.smoker_region_values = smoker_region_values
        self.smoker_charges_values = smoker_charges_values
    
    def average_smoker_age(self):
        smoker_total_age = 0
        for age in smoker_age_values:
            smoker_total_age += age
        smoker_average_age = smoker_total_age/len(smoker_age_values)  
        return "The average age for all smokers covered by this insurance is {}.".format(smoker_average_age)
    
    def smoker_sex_count(self):
        smoker_male_count = 0
        smoker_female_count = 0
        for sex in smoker_sex_values:
            if sex == "male":
                smoker_male_count +=1
            elif sex == "female":
                smoker_female_count+=1
        smoker_male_percentage = 100*(smoker_male_count/(smoker_male_count+smoker_female_count))
        smoker_female_percentage = 100-smoker_male_percentage
        return "Of all the smokers, {male}% are male whereas {female}% are female.".format(male=smoker_male_percentage, female=smoker_female_percentage)
    
    def smoker_bmi(self):
        smoker_total_bmi = 0
        for bmi in smoker_bmi_values:
            smoker_total_bmi += bmi
        smoker_average_bmi = smoker_total_bmi/len(smoker_bmi_values)
        number_overweight_smoker = 0
        for bmi in smoker_bmi_values:
            if bmi >= 25:
                number_overweight_smoker +=1
        percentage_overweight_smoker = 100*number_overweight_smoker/len(smoker_bmi_values)        
        return "The average bmi for all smokers is {bmi}, and {overweight}% of all the smokers are overweight.".format(bmi=smoker_average_bmi, overweight=percentage_overweight_smoker)        
    
    def average_smoker_children(self):
        smoker_total_children = 0
        for number_children in smoker_children_values:
            smoker_total_children += number_children
        smoker_average_children = smoker_total_children/len(smoker_children_values) 
        smoker_mode_children = statistics.mode(smoker_children_values)
        return "Smokers covered by this insurance have on average {a} children, while the number of children that appears the most is {b}.".format(a=smoker_average_children, b=smoker_mode_children)
    
    def smoker_region(self):
        num_southwest = 0
        num_southeast = 0
        num_northeast = 0
        num_northwest = 0
        for region in smoker_region_values:
            if region == "southwest":
                num_southwest +=1
            elif region == "southeast":
                num_southeast +=1
            elif region == "northeast":
                num_northeast +=1
            elif region == "northwest":
                num_northwest +=1
        region_count = {"Southwest":num_southwest, "Southeast":num_southeast, "Northeast":num_northeast, "Northwest":num_northwest}
        return region_count
                
        
        
    def smoker_charges(self):
        smoker_total_charges = 0
        for charge in smoker_charges_values:
            smoker_total_charges += charge
        smoker_average_charge = smoker_total_charges/len(smoker_charges_values) 
        smoker_charge_above_20k = 0
        for charge in smoker_charges_values:
            if charge > 20000:
                smoker_charge_above_20k +=1
        percentage_smoker_charge_above_20k = 100*smoker_charge_above_20k/len(smoker_charges_values)
        return "The average charges for all smokers is ${smoker_average_charge}, while {percentage_smoker_charge_above_20k}% of all smokers have a charge above $20k.".format(smoker_average_charge=smoker_average_charge, percentage_smoker_charge_above_20k=percentage_smoker_charge_above_20k )
            
            
```

```python
Smokers = SmokerAnalysis(smoker_age_values, smoker_sex_values, smoker_bmi_values, smoker_children_values, smoker_region_values, smoker_charges_values)
```
